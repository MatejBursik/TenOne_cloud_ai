```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file containing S&P 500 data
csv_filename = 'Data/SP500_last_decade.csv'
sp500_data = pd.read_csv(csv_filename, parse_dates=['Date'], index_col='Date')

# Plot the closing price over time
plt.figure(figsize=(10, 6))  # Set the figure size
plt.plot(sp500_data.index, sp500_data['Close'], label='S&P 500 Close', color='blue')

# Add title and labels
plt.title('S&P 500 Closing Prices Over the Last Year', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Closing Price (USD)', fontsize=12)

# Rotate the x-axis dates for better readability
plt.xticks(rotation=45)

# Add a grid
plt.grid(True)

# Add a legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# Load the CSV file containing S&P 500 data
sp500_data = pd.read_csv(csv_filename, parse_dates=['Date'], index_col='Date')

# Get the date one month ago
one_month_ago = datetime.now() - timedelta(days=30)

# Filter the data for the last month
last_month_data = sp500_data[sp500_data.index >= one_month_ago]

# Plot the daily opening prices
plt.figure(figsize=(10, 6))  # Set the figure size
plt.bar(last_month_data.index, last_month_data['Volume'], label='Volume', color='purple')

# Add title and labels
plt.title('S&P 500 Daily Volume', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Volume', fontsize=12)

# Rotate the x-axis dates for better readability
plt.xticks(rotation=45)

# Add a grid
plt.grid(True)

# Add a legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load CSV files for each MAMAA stock
csv_meta = 'Data/META_last_decade.csv'
csv_aapl = 'Data/AAPL_last_decade.csv'
csv_msft = 'Data/MSFT_last_decade.csv'
csv_amzn = 'Data/AMZN_last_decade.csv'
csv_googl = 'Data/GOOGL_last_decade.csv'
csv_sp500 = 'Data/SP500_last_decade.csv'

META_data = pd.read_csv(csv_meta, parse_dates=['Date'], index_col='Date')
AAPL_data = pd.read_csv(csv_aapl, parse_dates=['Date'], index_col='Date')
MSFT_data = pd.read_csv(csv_msft, parse_dates=['Date'], index_col='Date')
AMZN_data = pd.read_csv(csv_amzn, parse_dates=['Date'], index_col='Date')
GOOGL_data = pd.read_csv(csv_googl, parse_dates=['Date'], index_col='Date')
SP500_data = pd.read_csv(csv_sp500, parse_dates=['Date'], index_col='Date')

# Function to add a trendline
def add_trendline(ax, x, y, color='gray'):
    x_num = np.arange(len(x))  # Numeric index for fitting
    coeffs = np.polyfit(x_num, y, 1)  # Linear fit (degree 1)
    trendline = np.poly1d(coeffs)
    ax.plot(x, trendline(x_num), color=color, linestyle='--', linewidth=1)

# Set up the subplots with 3 rows and 2 columns
fig, axs = plt.subplots(3, 2, figsize=(16, 18), sharex=True)

# Plot Opening Prices for all MAMAA stocks on the left
axs[0, 0].plot(AAPL_data.index, AAPL_data['Open'], color='orange', label='AAPL Open')
add_trendline(axs[0, 0], AAPL_data.index, AAPL_data['Open'])
axs[0, 0].plot(META_data.index, META_data['Open'], color='blue', label='META Open')
add_trendline(axs[0, 0], META_data.index, META_data['Open'])
axs[0, 0].plot(MSFT_data.index, MSFT_data['Open'], color='green', label='MSFT Open')
add_trendline(axs[0, 0], MSFT_data.index, MSFT_data['Open'])
axs[0, 0].plot(AMZN_data.index, AMZN_data['Open'], color='red', label='AMZN Open')
add_trendline(axs[0, 0], AMZN_data.index, AMZN_data['Open'])
axs[0, 0].plot(GOOGL_data.index, GOOGL_data['Open'], color='purple', label='GOOGL Open')
add_trendline(axs[0, 0], GOOGL_data.index, GOOGL_data['Open'])
axs[0, 0].set_title('MAMAA Opening Prices Over the Last Year', fontsize=14)
axs[0, 0].set_ylabel('Open Price (USD)')
axs[0, 0].legend()
axs[0, 0].grid(True)

# Plot S&P 500 Opening Prices on the right
axs[0, 1].plot(SP500_data.index, SP500_data['Open'], color='black', label='S&P 500 Open')
add_trendline(axs[0, 1], SP500_data.index, SP500_data['Open'])
axs[0, 1].set_title('S&P 500 Opening Prices Over the Last Year', fontsize=14)
axs[0, 1].set_ylabel('Open Price (USD)')
axs[0, 1].legend()
axs[0, 1].grid(True)

# Plot Closing Prices for all MAMAA stocks on the left
axs[1, 0].plot(AAPL_data.index, AAPL_data['Close'], color='orange', label='AAPL Close')
add_trendline(axs[1, 0], AAPL_data.index, AAPL_data['Close'])
axs[1, 0].plot(META_data.index, META_data['Close'], color='blue', label='META Close')
add_trendline(axs[1, 0], META_data.index, META_data['Close'])
axs[1, 0].plot(MSFT_data.index, MSFT_data['Close'], color='green', label='MSFT Close')
add_trendline(axs[1, 0], MSFT_data.index, MSFT_data['Close'])
axs[1, 0].plot(AMZN_data.index, AMZN_data['Close'], color='red', label='AMZN Close')
add_trendline(axs[1, 0], AMZN_data.index, AMZN_data['Close'])
axs[1, 0].plot(GOOGL_data.index, GOOGL_data['Close'], color='purple', label='GOOGL Close')
add_trendline(axs[1, 0], GOOGL_data.index, GOOGL_data['Close'])
axs[1, 0].set_title('MAMAA Closing Prices Over the Last Year', fontsize=14)
axs[1, 0].set_ylabel('Close Price (USD)')
axs[1, 0].legend()
axs[1, 0].grid(True)

# Plot S&P 500 Closing Prices on the right
axs[1, 1].plot(SP500_data.index, SP500_data['Close'], color='black', label='S&P 500 Close')
add_trendline(axs[1, 1], SP500_data.index, SP500_data['Close'])
axs[1, 1].set_title('S&P 500 Closing Prices Over the Last Year', fontsize=14)
axs[1, 1].set_ylabel('Close Price (USD)')
axs[1, 1].legend()
axs[1, 1].grid(True)

# Plot Volume for all MAMAA stocks on the left
axs[2, 0].bar(AAPL_data.index, AAPL_data['Volume'], color='orange', label='AAPL Volume', width=1.0, alpha=0.5)
axs[2, 0].bar(META_data.index, META_data['Volume'], color='blue', label='META Volume', width=1.0, alpha=0.5)
axs[2, 0].bar(MSFT_data.index, MSFT_data['Volume'], color='green', label='MSFT Volume', width=1.0, alpha=0.5)
axs[2, 0].bar(AMZN_data.index, AMZN_data['Volume'], color='red', label='AMZN Volume', width=1.0, alpha=0.5)
axs[2, 0].bar(GOOGL_data.index, GOOGL_data['Volume'], color='purple', label='GOOGL Volume', width=1.0, alpha=0.5)
axs[2, 0].set_title('MAMAA Volume Over the Last Year', fontsize=14)
axs[2, 0].set_ylabel('Volume')
axs[2, 0].legend()
axs[2, 0].grid(True)

# Plot S&P 500 Volume on the right
axs[2, 1].bar(SP500_data.index, SP500_data['Volume'], color='black', label='S&P 500 Volume', width=1.0, alpha=0.5)
axs[2, 1].set_title('S&P 500 Volume Over the Last Year', fontsize=14)
axs[2, 1].set_ylabel('Volume')
axs[2, 1].legend()
axs[2, 1].grid(True)

# Set the x-axis label for the bottom plots
axs[2, 0].set_xlabel('Date', fontsize=12)
axs[2, 1].set_xlabel('Date', fontsize=12)

# Rotate x-axis dates for better readability
plt.xticks(rotation=45)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
