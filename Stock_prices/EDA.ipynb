```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file containing S&P 500 data
csv_filename = 'Data/SP500_last_year.csv'
sp500_data = pd.read_csv(csv_filename, parse_dates=['Date'], index_col='Date')

# Plot the closing price over time
plt.figure(figsize=(10, 6))  # Set the figure size
plt.plot(sp500_data.index, sp500_data['Close'], label='S&P 500 Close', color='blue')

# Add title and labels
plt.title('S&P 500 Closing Prices Over the Last Year', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Closing Price (USD)', fontsize=12)

# Rotate the x-axis dates for better readability
plt.xticks(rotation=45)

# Add a grid
plt.grid(True)

# Add a legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file containing S&P 500 data
sp500_data = pd.read_csv(csv_filename, parse_dates=['Date'], index_col='Date')

# Set up the subplots
fig, axs = plt.subplots(5, 1, figsize=(10, 15), sharex=True)

# Plot Opening Prices
axs[0].plot(sp500_data.index, sp500_data['Open'], color='orange', label='Open Price')
axs[0].set_title('S&P 500 Opening Prices Over the Last Year', fontsize=14)
axs[0].set_ylabel('Open Price (USD)')
axs[0].grid(True)

# Plot Closing Prices
axs[1].plot(sp500_data.index, sp500_data['Close'], color='blue', label='Close Price')
axs[1].set_title('S&P 500 Closing Prices Over the Last Year', fontsize=14)
axs[1].set_ylabel('Close Price (USD)')
axs[1].grid(True)

# Plot High Prices
axs[2].plot(sp500_data.index, sp500_data['High'], color='green', label='High Price')
axs[2].set_title('S&P 500 High Prices Over the Last Year', fontsize=14)
axs[2].set_ylabel('High Price (USD)')
axs[2].grid(True)

# Plot Low Prices
axs[3].plot(sp500_data.index, sp500_data['Low'], color='red', label='Low Price')
axs[3].set_title('S&P 500 Low Prices Over the Last Year', fontsize=14)
axs[3].set_ylabel('Low Price (USD)')
axs[3].grid(True)

# Plot Volume
axs[4].bar(sp500_data.index, sp500_data['Volume'], color='purple', label='Volume', width=1.0)
axs[4].set_title('S&P 500 Volume Over the Last Year', fontsize=14)
axs[4].set_ylabel('Volume')
axs[4].grid(True)

# Set the x-axis label
axs[4].set_xlabel('Date', fontsize=12)

# Rotate x-axis dates for better readability
plt.xticks(rotation=45)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
